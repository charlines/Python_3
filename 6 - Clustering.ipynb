# Clustering

## Importer le fichier

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "~/Documents/PDS/Cours/Data/bw_sample.csv"
train = pd.read_csv(path, sep=",",header=None)
cols = ["inf_2", "inf_7", "inf_15", "inf_30", "inf_50", "sup_50","mean_jx", "min_jx", 
        "max_jx", "nb_occ", "std","client_id", "mode_nb", "mode_jx"]
train.columns = cols

# Affichage des cinq premières lignes
train.head()

# Inf_2 = Nombre de voyages entre 0 et 2
# Inf_7 = Nombre de voyages entre 2 et 7
# Inf_15 = Nombre de voyages entre 7 et 15
```

```python
# Sélectionne les colonnes importantes et ajoute l'index
cols = ["client_id","inf_2", "inf_7", "inf_15", "inf_30", "inf_50", "sup_50","nb_occ","mean_jx", "mode_jx"]
df = train[cols]
df.set_index("client_id", inplace=True, drop=True)
df.head()
```

```python
col = df.apply(lambda row :(row["inf_2"]/row["nb_occ"]), axis = 1) 
df ["prop_inf_2"] = col
col = df.apply(lambda row :(row["inf_7"]/row["nb_occ"]), axis = 1) 
df ["prop_inf_7"] = col
col = df.apply(lambda row :(row["inf_15"]/row["nb_occ"]), axis = 1) 
df ["prop_inf_15"] = col
col = df.apply(lambda row :(row["inf_30"]/row["nb_occ"]), axis = 1) 
df ["prop_inf_30"] = col
col = df.apply(lambda row :(row["inf_50"]/row["nb_occ"]), axis = 1) 
df ["prop_inf_50"] = col
col = df.apply(lambda row :(row["sup_50"]/row["nb_occ"]), axis = 1) 
df ["prop_sup_50"] = col
df.head()
```

```python
cols = ["prop_inf_2", "prop_inf_7", "prop_inf_15", "prop_inf_30", "prop_inf_50", "prop_sup_50", "mean_jx", "mode_jx"]
df = df[cols]
```

```python
df.head()
```

## KMEANS

```python
# KMEANS
cols = ['prop_inf_2', 'prop_inf_7', 'prop_inf_15', 'prop_inf_30',
         'prop_inf_50', 'prop_sup_50']
X = df[cols]
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
```

```python
# Appliquer KMEANS
labels = kmeans.predict(X)
```

```python
# Récupérer les labels
df1 = df
df1['label'] = labels
```

```python
df1.head()
```

```python
df1.groupby("label").count()
```

```python
df1 = df1.groupby("label").mean()
df1
```

## GMM

```python
from sklearn.mixture import GaussianMixture
```

```python
cols = ['prop_inf_2', 'prop_inf_7', 'prop_inf_15', 'prop_inf_30',
         'prop_inf_50', 'prop_sup_50']
X = df[cols]
gmm = GaussianMixture(n_components=5, covariance_type='full').fit(X)
```

```python
labels = gmm.predict(X)
```

```python
df2 = df
df2['label'] = labels
```

```python
df2.head()
```

```python
df2.groupby("label").count()
```

```python
df2 = df2.groupby("label").mean()
df2
```
